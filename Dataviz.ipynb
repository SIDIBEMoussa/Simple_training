```
"Importation des librairies"
import matplotlib.pyplot as plt
"Création de variables"
x       = [1,2,3,4,5,6,7,8]
y       = [4,1,3,6,1,3,5,2]
y2     = [1, 4, 9, 16, 25,36,49,64]


"Création du graphique"

plt.title('la courbe sous Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, c='blue',label="courbe bleue")
plt.scatter(x, y2, c="green",label="courbe verte")
plt.legend()
plt.grid("True")
plt.annotate('mon commentaire', xy = (1, 4), xytext = (2, 22),arrowprops = {'facecolor': 'blue', 'shrink': 0.1})
plt.show()
```

```
# Importation des librairies
import numpy as np
import matplotlib.pyplot as plt

# Simulation des données
np.random.seed(19680801)

x1 = np.random.rand(50) * 100
x2 = np.ones(25) * 50
x_high = np.random.rand(10) * 100 + 100
x_low = np.random.rand(10) * -100
data = np.concatenate((x1, x2, x_high, x_low))

# Boxplot avec Matplotlib

plt.boxplot(data,vert=False)
plt.show()

plt.title('boxplot avec Matplotlib 2')
green_diamond = dict(markerfacecolor='b', marker='D')
plt.boxplot(data, flierprops=green_diamond,vert=False)
plt.show()
```
