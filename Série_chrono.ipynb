```
import matplotlib.pyplot as pl
import numpy as np
```

**Exercice 1:**

```
jours_str=["15 mars","16 mars","17 mars","18 mars", "19 mars"]
jours=[1,2,3,4,5]
cours=[135,143,140,154,152]
a=4.5
b=131.3
pl.plot(jours_str,cours,'bo-',label="observed")
jours.append(6)
y=[b+a*j for j in jours]
pl.rcParams.update({"font.size":16})
pl.title("L'evolution du cours")
jours_str.append("20 mars")
pl.plot(jours_str,y,'r>-',label="predicted")
pl.axes().set(xlabel="jours",ylabel="cours")
pl.annotate("tendance",xy=(3,149),xytext=(2.5,156),arrowprops = {'facecolor': 'red','shrink': 0.1})
pl.annotate("predicted value",xy=(1,144),xytext=(0.2,150),arrowprops = {'facecolor': 'blue','shrink': 0.1})
pl.xticks(jours_str,rotation=90)
pl.grid()
pl.legend()
pl.show()
```

**Exercice 2:**

```
str="235 298 221 340 268 327 242 378 300 368 292 421 334 421 322 465"
Zt=str.split(" ")
```

```
Zt=[int(i) for i in Zt]
t=range(1,17)
```

```
pl.plot(t,Zt,'bo-')
```

$Z_t=a t+b$ \\
$a=\frac{cov(Z_t,t)}{var(t)}$
b=$\overline Z_t-a \overline t$ \\
$\hat a=\frac{222}{7225} \qquad \hat b=\frac{138864}{425}$

```
pl.plot(t,Zt,'bo-')
a=222/7225
b=138864/425
Tendance=[a*j+b for j in t]
pl.plot(t,Tendance,"ro")
pl.show()
```

```
import pandas as pd
df=pd.DataFrame(columns=["t","Zt","Tendance"])
```

```
df.t=t;df.Zt=Zt;df.Tendance=Tendance
```
