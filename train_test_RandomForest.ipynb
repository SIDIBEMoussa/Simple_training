```
import pandas as pd
```

```
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
(train.shape,test.shape)
```

```
train.head(10)
```

```
import seaborn as sns
train.describe(include="all")
```

```
train=train.dropna()
```

```
test.describe(include="all")
```

```
test=test.dropna()
```

```
from sklearn.ensemble import RandomForestClassifier
import timeit
```

```
start_time=timeit.default_timer()
Forest=RandomForestClassifier(n_estimators=1000,oob_score=True)
```

```
x_train,y_train=train.iloc[:,:-2],train['Activity']
x_test,y_test=test.iloc[:,:-2],test['Activity']
```

```
x_test.shape
```

```
 Forest.fit(x_train,y_train)
```

```
y_pred=Forest.predict(x_test)
```

```
from sklearn.metrics import accuracy_score
acc1=accuracy_score(y_test,y_pred)
elapsed1=timeit.default_timer()-start_time
```

```
from sklearn.feature_selection import SelectFromModel
import timeit

start_time=timeit.default_timer()
model=SelectFromModel(Forest,prefit=True,threshold=0.003)
x_train2=model.transform(x_train)
x_train2.shape
```

```
Forest2=RandomForestClassifier(n_estimators=1000,oob_score=True).fit(x_train2,y_train)
```

```
x_train.shape
```

```
x_test.shape
```

```
x_test2=model.transform(x_test)
```

```
y_pred2=Forest2.predict(x_test2)
```

```
acc2=accuracy_score(y_test,y_pred2)
elapsed2=timeit.default_timer()-start_time
```

```
print("Accurarcy with features:{:.2%} time {:.2f}".format(acc1,elapsed1))
print("Accurarcy after selection with threshold:{:.2%} time {:.2f}".format(acc2,elapsed2))
```

On peut remarquer la difference de complexité qui se réfète nettement dans le temps d'exécution sans pour autant perdre beaucoup d'information qu'on peut estimer à travers le score de notre modèle de prédiction.

*   Cette étude n'est qu'un échantillon mais le temps la différence qui se présente ici peut être très  significatif dans la vie réelle alors cette astuce de distribution de l'importance avec threshold afin d'éffectuer une varselect sur les variables explicatives
