```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
import math
```

```python
def approxim_pi(N):
  M=np.random.rand(2,N)
  x,y=M[0][:],M[1][:]
  disque=x**2+y**2
  I=np.count_nonzero(disque<1)
  f=I/N
  pi=4*f
  return pi
```

```python
N=np.array([10,100,1000,10000,100000,1000000])
N=np.array(range(100,10000000,100000))
pi_array=[]
for n in N:
  pi_array.append(approxim_pi(n))
pi_array=np.array(pi_array)
```

```python
pl.style.use("ggplot")
pl.rcParams.update({"font.size":13})
pl.figure(figsize=(12,6))
pl.title('Simulation de la valeur de $\pi$')
fig=pl.plot(N,pi_array)
pl.plot(N,math.pi*np.ones(len(N)),"r--")
pl.axis(ymin=3.13,ymax=3.15)
pl.xlabel("Nombre de points")
pl.ylabel("Valeur de $\pi$")
```

En conclusion nous observons plus nous l'aire du disque est pleinement remplie plus on se rapproche de la valeur exacte de $\pi$
